### Separating X and Y data for Training Purpose

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
xtrain.shape
```

```python
150*30/100
```

```python
150*70/100
```

```python
ytrain
```

#### Evaluation Metrics

```python
from sklearn.metrics import classification_report
```

#### Creating functions for the Model

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    
    train = model.score(xtrain,ytrain)
    test = model.score(xtest,ytest)
    
    print(f"Training Accuracy:- {train}\n Testing Accuracy:- {test}")
    
    print(classification_report(ytest,ypred))
    return model
```

#### importing Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
logreg = mymodel(LogisticRegression())
```

```python
knn = mymodel(KNeighborsClassifier())
```

```python
SVC = mymodel(SVC())
```

```python
dt = mymodel(DecisionTreeClassifier())
```

### Lets try Scaling the value and then applying algorithms to it.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
logreg = mymodel(LogisticRegression())
```

```python
knn = mymodel(KNeighborsClassifier())
```

```python
SVC = mymodel(SVC)
```

```python
dt = mymodel(DecisionTreeClassifier())
```

#### Decision Tree Tuning

```python
parameter = {
            "criterion":["gini","entropy"],
             "max_depth":list(range(1,20)),
             "min_samples_leaf":list(range(1,20))
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(DecisionTreeClassifier(), parameter, verbose=2)
grid.fit(xtrain, ytrain)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
dt = mymodel(grid.best_estimator_)
```
